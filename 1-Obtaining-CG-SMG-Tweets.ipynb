# Obtaining CG & SMG Tweets

## 1. File history

The corpus of CG and SMG sentences was collected in March and April 2018. It was sourced from Twitter, Facebook, forums, and comments on blog posts. This notebook used to contain code to extract and save tweets by interacting with the Twitter API.

In 2022, Twitter was [bought by Elon Musk](https://en.wikipedia.org/wiki/Acquisition_of_Twitter_by_Elon_Musk), which brought sweeping changes to the platform. In February 2023, [Twitter announced](https://x.com/XDevelopers/status/1621026986784337922) that free access to the Twitter API would no longer be supported. Twitter was soon rebranded to “X”.

Following these changes, the code was removed from this notebook as it could no longer be run. It can still be viewed by accessing the file history on GitHub.

## 2. The collection process

Initially, using the Twitter API to search for Greek tweets by users located in Cyprus was tried as a method of finding CG tweets. This method proved inadequate since many of these were in SMG. Another approach was retrieving all tweets written by users who had previously tweeted in CG. However, this approach did not prove any better since it was observed that users who tweeted in CG may later tweet in SMG or even in English.

The best method for finding CG tweets was searching using **seed words**, while excluding retweets and reply tweets. Retweets were undesirable since they repeat the contents of other tweets that were already collected, while reply tweets often tended to be very short replies.

A list of 35 frequently-used words or phrases unique to the Cypriot dialect was compiled and used as seed words to search Twitter for CG tweets. An effort was made to limit the number of CG sentences collected in this manner in order not to introduce bias in the CG words represented in the corpus.
